## Movie Review Sentiment Analysis

We will be importing the imdb dataset, and assign it to two training sets as follows:

```python
from tensorflow.python.keras.datasets import imdb

(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words = 10000)
```

This is a single element in the training set:

```python
print(x_train[0])
```

We will be classifying the reviews as either negative or positive.

```python
class_names = ['Negative', 'Positive']
```

```python
word_index = imdb.get_word_index()
print(word_index['hello'])
```

## Decoding the Reviews

```python
reverse_word_index = dict((value, key) for key, value in word_index.items())

def decode(review):
    text = ''
    for i in review:
        text += reverse_word_index[i]
        text += ' '
    return text
```

Here, we will be decoding the set of numbers in the first element of the x_train training set. The sequence does not matter here.

```python
decode(x_train[0])
```

```python
def show_lengths():
    print('Length of 1st training example: ', len(x_train[0]))
    print('Length of 2nd training example: ',  len(x_train[1]))
    print('Length of 1st test example: ', len(x_test[0]))
    print('Length of 2nd test example: ',  len(x_test[1]))
    
show_lengths()
```

In the above lengths, we fix the total length as 256, which will be enough to contain all elements. So, to make all training and test sets the same size, we pad the training and test sets using the sequence function from keras.


## Padding the Examples

```python
word_index['the']
```

```python
from tensorflow.python.keras.preprocessing.sequence import pad_sequences

x_train = pad_sequences(x_train, value = word_index['the'], padding = 'post', maxlen = 256)
x_test = pad_sequences(x_test, value = word_index['the'], padding = 'post', maxlen = 256)
```

```python
show_lengths()
```

We can see from the above output that training and test sets have been padded to be equal in size.

```python
decode(x_train[0])
```

Here, we can see that the review has been padded with *"the"* to reach the 256 element limit.

## Creating and Training the Model

```python
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Embedding, Dense, GlobalAveragePooling1D

model = Sequential([
    Embedding(10000, 16),
    GlobalAveragePooling1D(),
    Dense(16, activation = 'relu'),
    Dense(1, activation = 'sigmoid')
])

model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['acc']
)

model.summary()
```

```python
from tensorflow.python.keras.callbacks import LambdaCallback

simple_logging = LambdaCallback(on_epoch_end = lambda e, l: print(e, end='.'))

E = 20

h = model.fit(
    x_train, y_train,
    validation_split = 0.2,
    epochs = E,
    callbacks = [simple_logging],
    verbose = False
)
```

## Predictions and Evaluation

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(range(E), h.history['acc'], label = 'Training')
plt.plot(range(E), h.history['val_acc'], label = 'Validation')
plt.legend()
plt.show()
```

From the above graph, we see that the training data performs extremely well, and seems to be highly accurate. However, this would be a case of overfitting, but since the validation curve lies between 85 to 90% accuracy, we can ignore it.

```python
loss, acc = model.evaluate(x_test, y_test)
print('Test set accuracy: ', acc * 100)
```

The accuracy of the model is givem as approximately 84%

```python
import numpy as np

prediction = model.predict(np.expand_dims(x_test[0], axis = 0))
class_names = ['Negative', 'Positive']
print(class_names[np.argmax(prediction[0])])
```
